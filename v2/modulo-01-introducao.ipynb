````python
# EXERC√çCIO 1: Simulando nosso primeiro LangChain app!
# (Vamos simular, pois n√£o temos API keys configuradas ainda)

class SimuladorLLM:
    """Simulador de LLM para fins educacionais"""
    
    def __init__(self, modelo="gemini-2.0-flash"):
        self.modelo = modelo
        self.historico = []
        
        # Respostas pr√©-definidas para demonstra√ß√£o
        self.respostas = {
            "o que √© langchain": "LangChain √© uma framework poderosa para construir aplica√ß√µes com LLMs, oferecendo abstra√ß√µes e ferramentas que simplificam o desenvolvimento de sistemas de IA conversacional.",
            "como funciona rag": "RAG (Retrieval-Augmented Generation) combina busca sem√¢ntica com gera√ß√£o de texto, permitindo que o modelo acesse informa√ß√µes espec√≠ficas de uma base de conhecimento para gerar respostas mais precisas.",
            "python ou javascript": "Python √© mais popular para IA/ML devido √†s suas bibliotecas robustas como TensorFlow, PyTorch e agora LangChain. JavaScript tem seu lugar, mas Python domina o ecossistema de IA.",
            "default": "Essa √© uma pergunta interessante! Com LangChain, podemos construir sistemas que respondem a qualquer tipo de quest√£o de forma inteligente e contextualizada."
    }
    
    def invoke(self, prompt):
        """Simula a chamada do LLM"""
        prompt_lower = prompt.lower()
        
        # Busca por palavras-chave
        for key in self.respostas:
            if key != "default" and key in prompt_lower:
                resposta = self.respostas[key]
                break
        else:
            resposta = self.respostas["default"]
        
        # Adiciona ao hist√≥rico
        self.historico.append({"pergunta": prompt, "resposta": resposta})
        
        return resposta

def criar_chat_simples():
    """Cria nosso primeiro 'chatbot' com LangChain (simulado)"""
    
    print("ü§ñ CHAT LANGCHAIN SIMULATOR v1.0")
    print("=" * 40)
    print("Perguntas sugeridas:")
    print("‚Ä¢ O que √© LangChain?")
    print("‚Ä¢ Como funciona RAG?")
    print("‚Ä¢ Python ou JavaScript?")
    print("‚Ä¢ Qualquer outra pergunta!")
    print("\n" + "=" * 40)
    
    # Inicializa nosso "LLM"
    llm = SimuladorLLM()
    
    # Algumas perguntas de demonstra√ß√£o
    perguntas_demo = [
        "O que √© LangChain?",
        "Como funciona RAG?",
        "Por que usar Python para IA?"
    ]
    
    print("üéØ DEMONSTRA√á√ÉO AUTOM√ÅTICA:\n")
    
    for i, pergunta in enumerate(perguntas_demo, 1):
        print(f"üë§ Usu√°rio: {pergunta}")
        resposta = llm.invoke(pergunta)
        print(f"ü§ñ LangChain: {resposta}\n")
        print("-" * 60 + "\n")
    
    return llm

# Executando nossa demonstra√ß√£o
chat = criar_chat_simples()

print("üìä ESTAT√çSTICAS DA SESS√ÉO:")
print(f"üí¨ Perguntas processadas: {len(chat.historico)}")
print(f"üß† Modelo simulado: {chat.modelo}")
print(f"‚ö° Tempo m√©dio de resposta: ~0.1s (simulado)")
print("\nüéâ Parab√©ns! Voc√™s acabaram de ver seu primeiro sistema LangChain em a√ß√£o!")
````

````python
# EXERC√çCIO DE PLANEJAMENTO INTERATIVO
# Vamos criar um framework para voc√™s planejarem sua aplica√ß√£o!

class PlanejadorLangChain:
    """Ferramenta para ajudar no planejamento de aplica√ß√µes LangChain"""
    
    def __init__(self):
        self.componentes = {
            'Models': {
                'descricao': 'LLMs para processamento de linguagem natural',
                'exemplos': ['GPT-4', 'Gemini', 'Claude', 'Llama'],
                'quando_usar': 'Sempre! √â o c√©rebro da aplica√ß√£o'
            },
            'Prompts': {
                'descricao': 'Templates para estruturar conversas',
                'exemplos': ['ChatPrompt', 'FewShotPrompt', 'SystemMessage'],
                'quando_usar': 'Para controlar o comportamento do LLM'
            },
            'Chains': {
                'descricao': 'Sequ√™ncias de opera√ß√µes conectadas',
                'exemplos': ['LLMChain', 'ConversationChain', 'SequentialChain'],
                'quando_usar': 'Para processos com m√∫ltiplos passos'
            },
            'Memory': {
                'descricao': 'Sistema para lembrar conversas anteriores',
                'exemplos': ['ConversationBufferMemory', 'ConversationSummaryMemory'],
                'quando_usar': 'Para chatbots e conversas longas'
            },
            'Documents': {
                'descricao': 'Processamento e carregamento de arquivos',
                'exemplos': ['PyPDFLoader', 'TextSplitter', 'CSVLoader'],
                'quando_usar': 'Para trabalhar com documentos externos'
            },
            'VectorStores': {
                'descricao': 'Bancos vetoriais para busca sem√¢ntica',
                'exemplos': ['Chroma', 'Pinecone', 'FAISS'],
                'quando_usar': 'Para sistemas RAG e busca inteligente'
            },
            'Agents': {
                'descricao': 'IA que toma decis√µes e usa ferramentas',
                'exemplos': ['ConversationalAgent', 'ReActAgent'],
                'quando_usar': 'Para automa√ß√£o complexa e tomada de decis√£o'
            }
        }
    
    def avaliar_projeto(self, descricao_projeto):
        """Sugere componentes baseado na descri√ß√£o do projeto"""
        descricao = descricao_projeto.lower()
        
        sugestoes = []
        
        # An√°lise simples por palavras-chave
        if any(word in descricao for word in ['chat', 'conversa', 'falar']):
            sugestoes.extend(['Models', 'Prompts', 'Memory'])
        
        if any(word in descricao for word in ['pdf', 'documento', 'arquivo', 'texto']):
            sugestoes.extend(['Documents', 'VectorStores'])
        
        if any(word in descricao for word in ['buscar', 'pesquisar', 'encontrar']):
            sugestoes.extend(['VectorStores', 'Agents'])
        
        if any(word in descricao for word in ['automatizar', 'decidir', 'escolher']):
            sugestoes.append('Agents')
        
        if any(word in descricao for word in ['passo', 'etapa', 'sequ√™ncia']):
            sugestoes.append('Chains')
        
        # Remove duplicatas mantendo ordem
        sugestoes_unicas = list(dict.fromkeys(sugestoes))
        
        # Sempre inclui Models e Prompts (b√°sicos)
        if 'Models' not in sugestoes_unicas:
            sugestoes_unicas.insert(0, 'Models')
        if 'Prompts' not in sugestoes_unicas:
            sugestoes_unicas.insert(1, 'Prompts')
        
        return sugestoes_unicas
    
    def gerar_relatorio(self, projeto, problema, componentes_sugeridos):
        """Gera um relat√≥rio completo do planejamento"""
        
        print("=" * 60)
        print(f"üéØ RELAT√ìRIO DE PLANEJAMENTO: {projeto.upper()}")
        print("=" * 60)
        
        print(f"\nüìù PROBLEMA IDENTIFICADO:")
        print(f"   {problema}")
        
        print(f"\nüîß COMPONENTES RECOMENDADOS:")
        for i, comp in enumerate(componentes_sugeridos, 1):
            info = self.componentes[comp]
            print(f"   {i}. {comp}")
            print(f"      üí° {info['descricao']}")
            print(f"      üéØ {info['quando_usar']}")
            print(f"      üì¶ Exemplos: {', '.join(info['exemplos'][:2])}...")
            print()
        
        # Estimativas
        complexidade = len(componentes_sugeridos)
        tempo_estimado = complexidade * 2  # 2 semanas por componente
        modulos_necessarios = min(complexidade + 2, 12)  # M√°ximo 12 m√≥dulos
        
        print(f"üìä ESTIMATIVAS:")
        print(f"   üéØ Complexidade: {complexidade}/7 componentes")
        print(f"   ‚è±Ô∏è Tempo estimado: {tempo_estimado} semanas")
        print(f"   üìö M√≥dulos necess√°rios: {modulos_necessarios}/15")
        
        valor_estimado = complexidade * 8000  # R$ 8k por componente
        print(f"   üí∞ Valor de mercado: R$ {valor_estimado:,}")
        
        return {
            'complexidade': complexidade,
            'tempo': tempo_estimado,
            'valor': valor_estimado,
            'componentes': componentes_sugeridos
        }

# Exemplos de projetos para demonstra√ß√£o
planejador = PlanejadorLangChain()

projetos_exemplo = [
    {
        'nome': 'Tutor Inteligente de Programa√ß√£o',
        'problema': 'Estudantes precisam de ajuda personalizada para aprender programa√ß√£o, com explica√ß√µes adaptadas ao seu n√≠vel e exemplos pr√°ticos baseados em documenta√ß√£o oficial.'
    },
    {
        'nome': 'Assistente Jur√≠dico Corporativo', 
        'problema': 'Advogados gastam muito tempo buscando informa√ß√µes em contratos e documentos legais, precisando de um sistema que encontre cl√°usulas espec√≠ficas e gere resumos autom√°ticos.'
    }
]

print("üöÄ EXEMPLOS DE PLANEJAMENTO:\n")

for exemplo in projetos_exemplo:
    componentes = planejador.avaliar_projeto(exemplo['problema'])
    resultado = planejador.gerar_relatorio(
        exemplo['nome'], 
        exemplo['problema'], 
        componentes
    )
    print("\n" + "="*60 + "\n")

print("\nüéØ AGORA √â SUA VEZ!")
print("Pense em um projeto pr√≥prio e use esta mesma estrutura!")
print("No final do curso, voc√™s ter√£o todas as ferramentas para implement√°-lo! üöÄ")
````
