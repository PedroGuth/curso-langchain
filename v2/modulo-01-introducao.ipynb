````python
# EXERCÃCIO 1: Simulando nosso primeiro LangChain app!
# (Vamos simular, pois nÃ£o temos API keys configuradas ainda)

class SimuladorLLM:
    """Simulador de LLM para fins educacionais"""
    
    def __init__(self, modelo="gemini-2.0-flash"):
        self.modelo = modelo
        self.historico = []
        
        # Respostas prÃ©-definidas para demonstraÃ§Ã£o
        self.respostas = {
            "o que Ã© langchain": "LangChain Ã© uma framework poderosa para construir aplicaÃ§Ãµes com LLMs, oferecendo abstraÃ§Ãµes e ferramentas que simplificam o desenvolvimento de sistemas de IA conversacional.",
            "como funciona rag": "RAG (Retrieval-Augmented Generation) combina busca semÃ¢ntica com geraÃ§Ã£o de texto, permitindo que o modelo acesse informaÃ§Ãµes especÃ­ficas de uma base de conhecimento para gerar respostas mais precisas.",
            "python ou javascript": "Python Ã© mais popular para IA/ML devido Ã s suas bibliotecas robustas como TensorFlow, PyTorch e agora LangChain. JavaScript tem seu lugar, mas Python domina o ecossistema de IA.",
            "default": "Essa Ã© uma pergunta interessante! Com LangChain, podemos construir sistemas que respondem a qualquer tipo de questÃ£o de forma inteligente e contextualizada."
    }
    
    def invoke(self, prompt):
        """Simula a chamada do LLM"""
        prompt_lower = prompt.lower()
        
        # Busca por palavras-chave
        for key in self.respostas:
            if key != "default" and key in prompt_lower:
                resposta = self.respostas[key]
                break
        else:
            resposta = self.respostas["default"]
        
        # Adiciona ao histÃ³rico
        self.historico.append({"pergunta": prompt, "resposta": resposta})
        
        return resposta

def criar_chat_simples():
    """Cria nosso primeiro 'chatbot' com LangChain (simulado)"""
    
    print("ğŸ¤– CHAT LANGCHAIN SIMULATOR v1.0")
    print("=" * 40)
    print("Perguntas sugeridas:")
    print("â€¢ O que Ã© LangChain?")
    print("â€¢ Como funciona RAG?")
    print("â€¢ Python ou JavaScript?")
    print("â€¢ Qualquer outra pergunta!")
    print("\n" + "=" * 40)
    
    # Inicializa nosso "LLM"
    llm = SimuladorLLM()
    
    # Algumas perguntas de demonstraÃ§Ã£o
    perguntas_demo = [
        "O que Ã© LangChain?",
        "Como funciona RAG?",
        "Por que usar Python para IA?"
    ]
    
    print("ğŸ¯ DEMONSTRAÃ‡ÃƒO AUTOMÃTICA:\n")
    
    for i, pergunta in enumerate(perguntas_demo, 1):
        print(f"ğŸ‘¤ UsuÃ¡rio: {pergunta}")
        resposta = llm.invoke(pergunta)
        print(f"ğŸ¤– LangChain: {resposta}\n")
        print("-" * 60 + "\n")
    
    return llm

# Executando nossa demonstraÃ§Ã£o
chat = criar_chat_simples()

print("ğŸ“Š ESTATÃSTICAS DA SESSÃƒO:")
print(f"ğŸ’¬ Perguntas processadas: {len(chat.historico)}")
print(f"ğŸ§  Modelo simulado: {chat.modelo}")
print(f"âš¡ Tempo mÃ©dio de resposta: ~0.1s (simulado)")
print("\nğŸ‰ ParabÃ©ns! VocÃªs acabaram de ver seu primeiro sistema LangChain em aÃ§Ã£o!")
````
